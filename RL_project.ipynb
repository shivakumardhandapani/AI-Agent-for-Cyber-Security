```python
vuln_info = explorer.analyze_vulnerabilities()

print("\n=== Node Vulnerabilities ===\n")
for node in vuln_info['node_vulnerabilities']:
    print(f"Node ID: {node['node_id']}")
    print(f"  Position: {node['position']}")
    print(f"  Defense Values: {node['defense_values']}")
    print(f"  Attack Values: {node['attack_values']}")
    print(f"  Detection: {node['detection']}")
    print(f"  Min Defense: {node['min_defense']}")
    print(f"  Max Defense: {node['max_defense']}")
    print("-" * 30)

print("\n=== Summary ===\n")
print(f"Vulnerabilities per Node: {vuln_info['vulnerabilities_per_node']}")
print(f"Vulnerabilities per Layer: {vuln_info['vulnerabilities_per_layer']}")
print(f"Total Vulnerabilities: {vuln_info['total_vulnerabilities']}")
```

```python
defense_stats = explorer.analyze_defense_effectiveness(num_episodes=10)

print("\n=== By Attack Type ===\n")
if defense_stats['by_attack_type']:
    for attack_type, values in defense_stats['by_attack_type'].items():
        print(f"- {attack_type}: {values}")
else:
    print("No data available.")

print("\n=== By Node ===\n")
for node_id, stats in defense_stats['by_node'].items():
    print(f"Node ID: {node_id}")
    print(f"  Total Defenses: {stats['total_defenses']}")
    print(f"  Successful Defenses: {stats['successful_defenses']}")
    print("-" * 30)

print("\n=== Detection Effectiveness ===\n")
if defense_stats['detection_effectiveness']:
    for index, effectiveness in enumerate(defense_stats['detection_effectiveness']):
        print(f"Detection {index + 1}: {effectiveness}")
else:
    print("No detection effectiveness data available.")
```

## SARSA algorithm

This notebook implements and trains a SARSA agent to defend against random attacks in the IDS Game environment.

### Import Libraries

```python
import os
import gymnasium as gym
import warnings
from datetime import datetime

from gym_idsgame.agents.training_agents.q_learning.q_agent_config import QAgentConfig
from experiments.util import util
from experiments.util.plotting_util import read_and_plot_results
from src.agents.sarsa_agent import SARSAAgent
from src.utils.utils import get_output_dir, print_summary
from src.environment.compatibility_wrapper import GymCompatibilityWrapper
from src.utils.plotting import plot_results

%matplotlib inline
warnings.filterwarnings('ignore')
```

### Setup and Configuration

```python
random_seed = 33
output_dir = get_output_dir(algorithm='sarsa')
util.create_artefact_dirs(output_dir, random_seed)

sarsa_config = QAgentConfig(
    gamma=0.999,                # Discount factor
    alpha=0.0005,              # Learning rate
    epsilon=1,                 # Initial exploration rate
    min_epsilon=0.01,          # Minimum exploration rate
    epsilon_decay=0.9999,      # Decay rate per episode

    # Training parameters
    num_episodes=20001,
    train_log_frequency=100,
    eval_log_frequency=1,
    eval_frequency=1000,
    eval_episodes=100,

    # Agent type
    attacker=False,            # No Attacker --> BOT
    defender=True,             # This Agent is a Defender

    # Logging e savings
    render=False,
    video=False,
    video_frequency=101,
    eval_render=False,
    gifs = False,
    video_fps=5,
    video_dir=output_dir + "/results/videos/" + str(random_seed),
    gif_dir=output_dir + "/results/gifs/" + str(random_seed),
    save_dir=output_dir + "/results/data/" + str(random_seed)
)

print("Configuration initialized!")
print(f"Output directory: {output_dir}")
```

### Environment Setup

```python
env_name = "idsgame-random_attack-v8"
env = gym.make(env_name, save_dir=output_dir + "results/data/" + str(random_seed))
env = GymCompatibilityWrapper(env)

print("\nEnvironment Information:")
print(f"Observation Space: {env.observation_space}")
print(f"Action Space: {env.action_space}")
print(f"Defender Action Space: {env.defender_action_space}")
```

### Training

```python
sarsa_agent = SARSAAgent(env, sarsa_config)

print("Starting training...")
train_result, eval_result = sarsa_agent.train()
print("Training completed!")
```

### Results Analysis

```python
train_path = sarsa_config.save_dir + "/1732793675.598363_train_results_checkpoint.csv"
eval_path = sarsa_config.save_dir + "/1732793675.598363_eval_results_checkpoint.csv"

read_and_plot_results(
    train_path, eval_path,
    sarsa_config.train_log_frequency,
    sarsa_config.eval_frequency, sarsa_config.eval_log_frequency,
    sarsa_config.eval_episodes, output_dir, sim = False,
    random_seed = random_seed)
```

```python
#output_dir = ""
random_seed = 33
plot_results(output_dir, random_seed, algorithm="SARSA", mode='random')
```

```python
print_summary(train_result, title = "Training")
print("\n")
print_summary(eval_result, title = "Evaluation")
```

### Save final results

```python
os.makedirs("final_results", exist_ok=True)

timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
results_path = os.path.join("final_results/", f"final_results_{timestamp}")

train_result.to_csv(results_path + "_train.csv")
eval_result.to_csv(results_path + "_eval.csv")

print(f"\nResults saved to: {results_path}")
```

## DDQN algorithm

### Import Libraries

```python
import os
import gymnasium as gym
import warnings
from datetime import datetime

from gym_idsgame.agents.training_agents.q_learning.q_agent_config import QAgentConfig
from gym_idsgame.agents.training_agents.q_learning.dqn.dqn_config import DQNConfig
from experiments.util import util
from experiments.util.plotting_util import read_and_plot_results
from src.agents.ddqn_agent import DDQNAgent
from src.utils.utils import get_output_dir, print_summary
from src.environment.compatibility_wrapper import GymCompatibilityWrapper
from src.utils.plotting import plot_results

warnings.filterwarnings('ignore')
```

### Setup and Configuration

```python
random_seed = 33
output_dir = get_output_dir(algorithm='ddqn')
util.create_artefact_dirs(output_dir, random_seed)

dqn_config = DQNConfig(
    input_dim=33,
    attacker_output_dim=33,
    defender_output_dim=33,
    hidden_dim=64,
    replay_memory_size=1000,
    num_hidden_layers=1,
    replay_start_size=100,
    batch_size=32,
    target_network_update_freq=100,
    gpu=True,
    tensorboard=False,
    optimizer="Adam",
    lr_exp_decay=True,
    lr_decay_rate=0.999
)

ddqn_config = QAgentConfig(
    # Learning parameters
    gamma=0.9,
    alpha=0.0001,
    epsilon=1,
    min_epsilon=0.01,
    epsilon_decay=0.999,

    # Training parameters
    num_episodes=5000,
    train_log_frequency=100,
    eval_log_frequency=1,
    eval_frequency=250,
    eval_episodes=100,
    checkpoint_freq=1000,

    # Agent type
    attacker=False,
    defender=True,

    # Logging and saving
    render=False,
    video=False,
    video_frequency=101,
    eval_render=False,
    gifs=False,
    video_fps=5,
    video_dir=output_dir + "/results/videos/" + str(random_seed),
    gif_dir=output_dir + "/results/gifs/" + str(random_seed),
    save_dir=output_dir + "/results/data/" + str(random_seed),

    dqn_config=dqn_config
)
```

```python
# DQN specific parameters

ddqn_config.dqn_config.input_dim = 33
ddqn_config.dqn_config.output_dim = 30
ddqn_config.dqn_config.hidden_dim = 64
ddqn_config.dqn_config.replay_memory_size = 1000
ddqn_config.dqn_config.batch_size = 32
ddqn_config.dqn_config.target_network_update_freq = 100
ddqn_config.dqn_config.gpu = True
ddqn_config.dqn_config.tensorboard_dir = output_dir + "/results/tensorboard/" + str(random_seed)

print("Configuration initialized!")
print(f"Output directory: {output_dir}")
```

### Environment Setup

```python
env_name = "idsgame-maximal_attack-v8"
env = gym.make(env_name, save_dir=output_dir + "results/data/" + str(random_seed))
env = GymCompatibilityWrapper(env)

print("\nEnvironment Information:")
print(f"Observation Space: {env.observation_space}")
print(f"Action Space: {env.action_space}")
print(f"Defender Action Space: {env.defender_action_space}")
```

### Training

```python
ddqn_agent = DDQNAgent(env, ddqn_config)

print("Starting training...")
train_result = ddqn_agent.train()
print("Training completed!")
```

### Results Analysis

```python
train_path = "results/training/ddqn_2024-12-03_14-13-23/results/data/33/1733231094.9924657_train_results_checkpoint.csv"
eval_path = "results/training/ddqn_2024-12-03_14-13-23/results/data/33/1733231094.9924657_eval_results_checkpoint.csv"

read_and_plot_results(
    train_path, eval_path,
    ddqn_config.train_log_frequency,
    ddqn_config.eval_frequency,
    ddqn_config.eval_log_frequency,
    ddqn_config.eval_episodes,
    output_dir,
    sim=False,
    random_seed=random_seed
)
```
